```python
import leafmap
```

```python
m = leafmap.Map ()
m
```

```python
m.user_roi
```

```python
point_geometry = {"x": -82.229919, "y": 33.69578}
```

```python
gdf = leafmap.get_wbd(point_geometry, digit=8, return_geometry=True)
gdf.explore()
```

```python
m.user_roi_bounds()
```

```python
bbox_geometry = {"xmin": -82.295, "ymin": 33.664, "xmax": -82.163, "ymax": 33.7386}
```

```python
gdf = leafmap.get_wbd(bbox_geometry, digit=8, return_geometry=True)
gdf.explore()
```

```python
huc_id = "10160002"
gdf = leafmap.get_wbd(searchText=huc_id, digit=8)
gdf.explore()
```

```python
gdf = leafmap.get_wbd(searchText=huc_id, digit=10)
gdf.explore()
```
